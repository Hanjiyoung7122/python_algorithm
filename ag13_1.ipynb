```python
nums = [1, 3, 5, 7, 9, 10, 20, 10]

# 1번째 부터 8번째 까지
# 인덱스 0부터 7까지

t1 = 0
for i in range(8): # 0~7
    t1 = t1 + nums[i]

print(t1)


# 1번째 부터 3번째 까지
t2 = 0
for i in range(3): # 0 ~ 2
    t2 = t2 + nums[i]
print(t2)

```

```python
nums = [1, 3, 5, 7, 9, 10, 20, 10]
#누적합 0부터 시작
prefix = [0]
for n in nums:
    # 0 + n
    # prefix[1] + 3
    # prefix[2] + 5
    # ...
    prefix.append(prefix[-1]+n)
print(nums)
print(prefix)
```

```python
# 1 처음부터 8 끝까지 -> 인덱스 : 0~ 7까지
prefix[8]
```

```python
# 1부터 7까지 -> 인덱스 0부터 6까지
prefix[7]
```

```python
# 원본의 5번째에서 7번째까지 -> 인덱스 4 ~ 6
s = 5
e = 7
print(prefix[e] - prefix[s-1])
```

```python
scores = [10, 20, 3, 40, 5, 60, 8, 55, 44]

k = 3 # window size

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[0:k]) #score[:k]

totals = [window]
avgs = [window / k]
max_val = {"구간":f"0~2", "합":window}
print("max_val", max_val)
min_val = {"구간":f"0~2", "합":window}

print(window) #60

for i in range(k, len(scores)): # range(3, 6) 3zks dlehd / 원도우크기 부터 1칸식 이동하며 윈도우안의 합을 연산
    window = window + scores[i] - scores[i-k]
    totals.append(window)
    avgs.append(window/k)

    section = f"{i-(k-1)} ~ {i}"

    if window > max_val["합"]:
        max_val = {"구간": section,"합":window}

    if window < min_val["합"]:
        min_val = {"구간": section,"합":window}

    # print(window)

    # window + scores[3] - scores[3-3] -> 90
    # window + scores[4] - scores[4-3] -> 120
    # window + scores[5] - scores[5-3] -> 150

    print(totals)
    print(avgs)
    print(max_val)
    print(min_val)
```
