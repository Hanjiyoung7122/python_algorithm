```python
# 숫자 문자 제거
import re

text = "A10 B20 C30"
# result = re.sub(r"\d", "", text)
result = re.sub(r"[^\d\s]", "0", text) # #공백도 치환됨으로 공백을 빼는 방법도 아는것이 좋다
print(result)
```

```python
import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    name = re.search(r"[가-힣]", product)
    price = re.search(r"\d+", product)
    #print(price.group())
    price_list.append({name.group():price.group()})

print(price_list)
# [{"사과":3500}, {"배":4200}, ...]
```

```python
import re

emails = [
    "python@네이버.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```

```python
import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# ^https? -> "http" or "https" 로 시작
for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
