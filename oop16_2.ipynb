```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def greet(self):
        print(f"{self.name}님 hello")

class Student2:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def study(self):
        print("공부한다.")

class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
s1 = Student("뷔")
s1.print_name()
s1.greet()

s2 = Student2("지민")
s2.print_name()
s2.study()
```

```python
s = HighStudent("홍길동")
s.print_name()
s.study()
```

```python
def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))
```

```python
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = 0



    def average(self):
        return (self.kor + self.eng + self.math) / 3


s = Student("홍길동", 90, 85)
s.math = 88
s2 = Student("김철수", 50, 60)
s2.math = 70
print(s.average())
print(s2.average())
```
