```python
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 
```

```python
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
import re

text = "Python"

print(re.search(r"hon$", text)) # <re.Match object; span=(3, 6), match='hon'>
print(re.search(r"abc$", text)) # None
```

```python
import re

text = "Pthon"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py+", text)) # <re.Match object; span=(0, 6), match='Pyyyyy'>
```

```python
import re

texts = ["123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d* : \d+ : 숫자가 1개 이상 있어야 한다.
```

```python
import re

text = "Py@thon 123_xyz한글"

print(re.findall(r"\d", text))
print(re.findall(r"\D", text))
print(re.findall(r"\w", text))
print(re.findall(r"\W", text)) # 특수문자, 공백
print(re.findall(r"\s", text)) # 공백
```
