```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count

    def print_info(self):
        print(self.name)
        print(self.price)
        print(self.stock)

# 세일즈 목록
sales = [
    Product("키보드", 50000, 10),
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3)
]

#{"키보드" : 2, "마우스" : 1, "모니터" : 1, "노트북" : 3}
counts = {}
for s in sales:
    if s.name not in counts: #counts에 상품이름 키가 없다면
        counts[s.name] = 1
    else:
         counts[s.name] = counts[s.name] + 1 #counts[s.name] += 1

print(counts)

check_name_list = [] # 같은 상품이름 2번 계산되면 안되니까 계산했으면 입력
for s in sales:
    if s.name not in check_name_list:
        # count = counts[s.name]
        print(f"{s.name} : {s.total_price(counts[s.name])}") # 키보드 : s.total_price
        check_name_list.append(s.name)
```
