```python
steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]

total = 0

for i in range(len(steps)):
    total = total + steps[i]
    print(f"{i+1}일 누적: {total}")

```

```python
steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

total = 0

for day in steps:
    for date, count in day.items():
        total = total + count
        print(f"{date} 누적 : {total}")
```
