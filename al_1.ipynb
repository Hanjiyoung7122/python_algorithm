```python
# 1
num = int(input("학생 번호: "))
#num

#(num-1) // 10
#1~10 -> 0
#11~20 -> 10
leader_num = ((num-1) // 10 * 10) + 1

print(leader_num)
```

```python
# 2
try:
    num = int(input("숫자를 입력하세요"))

    if num % 2 == 0:
        print("짝수")
    else:
        print("홀수")
except ValueError as e:
    print("숫자만 입력하세요")
    #print(e)
```

```python
try:
    num = int(input("정수를 입력하세요: "))

    if num > 0:
        print("양수")
    elif num < 0:
        print("음수")
    else:
        print(0)
except ValueError as e:
    print("숫자만 입력하세요")
    
```
